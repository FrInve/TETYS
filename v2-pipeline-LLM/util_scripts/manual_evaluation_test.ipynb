```python
GROUP=2
CONFIG=2
```

```python
df = pd.read_parquet(f'../data/processed/group_{GROUP}/test_data_2024_random_400_clean_group_{GROUP}.parquet')
abstracts = df.abstract.to_list()
len(abstracts)
```

```python
embed = f'../data/processed/group_{GROUP}/embeddings_random_400_LLM_test_data_2024_group_{GROUP}.npy'
with open(embed,'rb') as f:
    test_embeddings = np.load(f)
```

```python
len(test_embeddings)
```

```python
# topic_model_config_1 = BERTopic.load(f"../models/group_{GROUP}/model_group_{GROUP}_dbcv_0_72", embedding_model="sentence-transformers/allenai-specter")
topic_model_config_2 = BERTopic.load(f"../models/group_{GROUP}/random_LLM_model_group_{GROUP}_dbcv_0_76", embedding_model="Salesforce/SFR-Embedding-2_R")
```

```python
num_of_abs= len(topic_model_config_2.topics_)
(topic_model_config_2.topic_sizes_[-1]/num_of_abs)*100
```

```python
# predicted_topics = topic_model_config_1.transform(abstracts)
predicted_topics = topic_model_config_2.transform(abstracts, embeddings=test_embeddings)
```

```python
df["topic_id"] = predicted_topics[0]
df["topic_prob"] = predicted_topics[1]
df[df.topic_id != -1]
```

```python
def print_info(position,topic_model,predicted_topics):
    predicted = predicted_topics[0][position]
    print("----")
    print(df.abstract.iloc[position])
    print(topic_model.get_topic_info(predicted).Name[0])
    print(topic_model.get_topic_info(predicted).Count[0])
    print(topic_model.get_topic(predicted))
```

```python
print_info(46, topic_model_config_2, predicted_topics)
```

```python
df["topic_name"] = df.topic_id.map(lambda x: topic_model_config_2.get_topic_info(x).Name[0])
df["topic_count"] = df.topic_id.map(lambda x: topic_model_config_2.get_topic_info(x).Count[0])
df["topic_terms"] = df.topic_id.map(lambda x: topic_model_config_2.get_topic(x))
```

```python
df.to_excel(f"../sdg_evaluation_2024_random_400/group_{GROUP}/articles_group_{GROUP}_config_{CONFIG}_all.xlsx",header=True)
df[df.topic_id != -1][["prism_doi", "dc_title", "abstract", "authors", "prism_cover_date", "author_keywords", "subject_area", "topic_id", "topic_prob", "topic_name", "topic_terms"]].to_excel(f"../sdg_evaluation_2024_random_400/group_{GROUP}/articles_test_group_{GROUP}_config_{CONFIG}_no_outliers.xlsx",header=True, index=False)
```
